## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from sklearn.preprocessing import OneHotEncoder

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[["Attrition","Department"]]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data

X_list = ["Education","Age","DistanceFromHome","JobSatisfaction","OverTime","StockOptionLevel","WorkLifeBalance","YearsAtCompany","YearsSinceLastPromotion","NumCompaniesWorked" ]

# Create X_df using your selected columns

X = attrition_df[X_list]

# Show the data types for X_df

X.dtypes

```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y_df, random_state=78)
```

```python
X_train['OverTime'].value_counts()
```

```python
y_df
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

# OneHotEncode the OverTime column for X
columns_to_encodeX = ["OverTime"]

oh2_Xtr = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
oh2_Xtr.set_output(transform="pandas")

# Fit and transform the OneHotEncoder to the columns to encode
X_train_encoded = oh2_Xtr.fit_transform(X_train[columns_to_encodeX])
X_test_encoded = oh2_Xtr.transform(X_test[columns_to_encodeX])

```

```python
X_test_encoded
```

```python
# Merge one-hot encoded features and drop the originals
X_train = X_train.merge(X_train_encoded,left_index=True, right_index=True)
X_train = X_train.drop(columns_to_encodeX, axis=1)
X_train.head()
```

```python
# Merge one-hot encoded features and drop the originals
X_test = X_test.merge(X_test_encoded,left_index=True, right_index=True)
X_test = X_test.drop(columns_to_encodeX, axis=1)
X_test.head()
```

```python
# Create a StandardScaler

scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
# Create a OneHotEncoder for the Department and Attrition columns
# Fit the encoder to the training data
# Create two new variables by applying the encoder
# to the training and testing data

columns_to_encode_y = ["Department","Attrition"]

ohe_ytr = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ohe_ytr.set_output(transform="pandas")

# Fit and transform the OneHotEncoder to the columns to encode
y_train_encoded = ohe_ytr.fit_transform(y_train[columns_to_encode_y])
y_test_encoded = ohe_ytr.transform(y_test[columns_to_encode_y])

```

```python
y_test_encoded
```

```python
X_train.shape[1]
```

## Create, Compile, and Train the Model

```python
# Create the shared layers of the model

# Input layer
input_layer = layers.Input(shape=(X_train.shape[1],), name='input_features')

# Shared hidden layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(128, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_dense = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
department_output = layers.Dense(3, activation='softmax',name='department_output')(department_dense)
#chamge back to softmax, testing
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer

attrition_dense = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
attrition_output = layers.Dense(2, activation='softmax',name='attrition_output')(attrition_dense)
```

```python


# Create the model
model = Model(inputs=input_layer, outputs=[department_output,attrition_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'department_output': 'categorical_crossentropy',
                     'attrition_output': 'categorical_crossentropy'},
              metrics={'department_output': 'accuracy',
                        'attrition_output': 'accuracy'})
# Summarize the model
model.summary()
#change attrition to binary_crossentropy
```

```python
y_train_department = y_train_encoded[["Department_Human Resources","Department_Research & Development","Department_Sales"]]
y_train_attrition = y_train_encoded[["Attrition_No","Attrition_Yes"]]
```

```python
y_test_department = y_test_encoded[["Department_Human Resources","Department_Research & Development","Department_Sales"]]
y_test_attrition = y_test_encoded[["Attrition_No","Attrition_Yes"]]
```

```python
# Train the model
model.fit(
    X_train_scaled,
    {
        'department_output': y_train_department,
        'attrition_output': y_train_attrition
    },
    epochs=100,  # You can adjust the number of epochs based on your needs
    batch_size=1,  # You can adjust the batch size based on your available memory
    validation_data=(X_test_scaled, y_test_department, y_test_attrition) # You can specify the validation split if you have a separate validation set
)
#tried other combinations for epochs and batch size, but did not see much improvement
```

```python
results = model.evaluate(np.array(X_test_scaled), {
        'department_output': y_test_department,
        'attrition_output': y_test_attrition   
})
```

```python
print(f"Results: {results}")
```

```python
print(f"Department predictions accuracy: {results[2]}")
print(f"Attrition predictions accuracy: {results[1]}")
```

## Answer (1):

Given the distribution, we see it is an imbalanced training dataset for both department and attrition. 
- Count by department in training set is 44, 734, 324
- Count by attrition in training set is 912, 190

In such cases, accuracy is not the best metric. Better metrics may be Precision, Recall, F1-score etc.

## Answer (2):
We have a multi-class classification. 
- Department is a 3-class classification and can have 3 values. But each instance can belong to exactly one class out of 3.
- Attrition is a 2-class classification and can have 2 values. Here also each instance can belong to exactly one class out of 2.

In both cases I have selected softmax activation function for this reason. If an instance can simultaneously belong to more than one class, then sigmoid is better fit. 
Note: In the case of Attrition, it is a binary classification and sigmoid can be selected as well.
